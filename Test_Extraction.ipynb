### Tests if keywords are being scraped and if scraped domains are being excluded from master list

```python
import os
import pandas as pd
```

```python
unparsed_path=os.getcwd()+"\\unscraped\\pickled_cats_cleaned.pkl"
parsed_path=os.getcwd()+"\\scraped\\"
```

```python
df_unparsed=pd.read_pickle(unparsed_path).reset_index()
parsed_files=os.listdir(parsed_path)
df_parsed=pd.concat([pd.read_pickle(parsed_path+file) for file in parsed_files]).rename(columns={'site': 'url_domain'})
df_parsed.head()
```

```python
pd.merge(df_unparsed, df_parsed, on='url_domain')
```
